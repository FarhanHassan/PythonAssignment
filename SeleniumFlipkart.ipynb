```python
# Install necessary libraries and drivers
!pip install selenium webdriver-manager > /dev/null 2>&1
!apt-get update > /dev/null 2>&1
!apt-get install -y chromium-chromedriver > /dev/null 2>&1

# Import required modules
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

# Configure Chrome options for Colab
options = Options()
options.add_argument('--headless')  # Run in headless mode
options.add_argument('--no-sandbox')  # Disable sandbox
options.add_argument('--disable-dev-shm-usage')  # Overcome resource limits
options.add_argument('--disable-gpu')  # Optional for better compatibility

# Use the correct path for the ChromeDriver installed in Colab
chrome_driver_path = '/usr/bin/chromedriver'

# Initialize the WebDriver
driver = webdriver.Chrome(options=options)

# Automate Flipkart search
try:
    # Open Flipkart
    driver.get("https://www.flipkart.com/")
    print("Page Title:", driver.title)

    # Handle Flipkart login popup
    try:
        close_button = driver.find_element(By.XPATH, "//button[contains(text(),'✕')]")
        close_button.click()
    except Exception as e:
        print("Login popup not displayed:", e)

    # Search for "one plus headphones"
    search_box = driver.find_element(By.XPATH, "//input[@type='text']")
    search_box.send_keys("one plus headphones")
    search_box.submit()

    # Wait for search results and interact with the first result
    driver.implicitly_wait(10)
     # Find all elements with the class "wjcEIp"
    elements = driver.find_elements(By.CLASS_NAME, "wjcEIp")

# Extract href and title for each element
    links_and_titles = [{"url": element.get_attribute("href"), "title": element.text} for element in elements]

# Find the image matching the given CSS selector (assuming only one image)
    image = driver.find_element(By.CSS_SELECTOR, ".VJA3rP > div > div > div._4WELSP > img.DByuf4")
    image_src = image.get_attribute("src")

# Find the single price using the given CSS selector (assuming one price)
    price = driver.find_element(By.CSS_SELECTOR, "a>.hl05eU>div.Nx9bqj").text

# Start printing links, titles, prices, and image URLs in the same loop
    for i, item in enumerate(links_and_titles, start=1):
      print(f"Link {i}: {item['url']}")
      print(f"Title {i}: {item['title']}")
      print(f"Price: {price}")  # The same price for each item
      print(f"Image {i}: {image_src}")  # Assuming only one image for all items

# Display the current URL
    print("Current URL:", driver.current_url)

finally:
    # Close the browser
    driver.quit()
```
